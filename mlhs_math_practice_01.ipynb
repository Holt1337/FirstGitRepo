```python
# your code here

import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x ** 2 - 30 * x + 1

def g(x):
    return -x ** 3 - 7 * x ** 2 + 5 * x + 17
def h(x):
    return x ** 2 - 1 / x
def j(x):
    return 1 / x ** 5 + x ** 17 +8
    
def search_extremum_point(fun):
    df_x = sp.diff(fun(x))
    print('Function: ', fun(x))
    extremums = sp.solve(df_x, x)
    if len(extremums) == 0:
        print('No extremum points')
        print()
        return extremums
    print('Extremum points:')
    for extremum in extremums:
        print('\t(', extremum, ', ', fun(extremum), ')', sep='')
    print()
    return extremums

if __name__ == '__main__':
    functions = [f, g, h, j]
    x = sp.Symbol('x', real=True)
    for fun in functions:
        search_extremum_point(fun)
```

## Задание 4

Найдите точки перегиба этих функций в Python.

```python
# your code here
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x ** 2 - 30 * x + 1

def g(x):
    return -x ** 3 - 7 * x ** 2 + 5 * x + 17
def h(x):
    return x ** 2 - 1 / x
def j(x):
    return 1 / x ** 5 + x ** 17 +8

def search_inflaction_point(fun):
    df_x = sp.diff(fun(x))
    d2f_x = sp.diff(df_x)
    print('Function: ', fun(x))
    inflactions = sp.solveset(d2f_x, x, domain=sp.S.Reals)
    if len(inflactions) == 0:
        print('No inflaction points')
        print()
        return inflactions
    print('Inflaction points:')
    for inflaction in inflactions:
        print('\t(', inflaction, ', ', fun(inflaction), ')', sep='')
    print()
    return inflactions

if __name__ == '__main__':
    functions = [f, g, h, j]
    x = sp.Symbol('x', real=True)
    for fun in functions:
        search_inflaction_point(fun)
```

## Задание 6

Модифицируйте функцию, осуществляющую метод градиентного спуска с семинара: добавьте опцию останавливать метод, если абсолютное значение производной функции в точке < некоторого $\delta$ (delta добавьте в аргументы функции).

```python
# your code here
def gradient_descent(x_start, learning_rate, epsilon, num_iterations, delta, fun):
    x_curr = x_start
    df_x = sp.diff(fun(x)) # производная

    trace = []
    trace.append(x_curr)

    for i in range(num_iterations):
      x_new = x_curr - learning_rate * df_x.subs(x, x_curr) 
      trace.append(x_new)
      if abs(df_x.subs(x, x_new)) < delta:
        return x_curr, trace, i+1

      if abs(x_new - x_curr) < epsilon:
        return x_curr, trace, i+1

      x_curr = x_new

    return x_curr, trace,  num_iterations
```

## Задание 8

Реализуйте [метод Ньютона](https://translated.turbopages.org/proxy_u/en-ru.ru.c97ef85d-631c4b6a-ff938028-74722d776562/https/en.wikipedia.org/wiki/Newton%27s_method_in_optimization) (метод второго порядка) для поиска минимума функции.

Про метод ещё можно почитать [здесь](https://math.semestr.ru/optim/minimum.php).

```python
# your code here
def NewtonMethod(fun, x_start, x_end, epsilon, num_iterations):

    df_x = sp.diff(fun(x))
    df2_x = sp.diff(df_x)
    df3_x = sp.diff(df2_x)
    x_cur = x_start if df_x.subs(x, x_start)*df3_x.subs(x, x_start) > 0 else x_end
    print('Начальное значение точки: ', x_cur)
    trace = []
  
    for i in range(num_iterations):
        x_new = x_cur - df_x.subs(x, x_cur)/df2_x.subs(x, x_cur)
        trace.append(x_new)
        if abs(x_cur - x_new) < epsilon:
            return x_new, trace, i+1
        x_cur = x_new
    return x_cur, trace, num_iterations+1
```

## Задание 11*

Реализуйте оптимизацию метода градиентного спуска - [Momentum](https://ml-handbook.ru/chapters/optimization/intro#метод-инерции-momentum).

```python
# your code here
def Momentum(fun, x_start, alpha, betta, epsilon, num_iterations):
    x_curr = x_start
    df_x = sp.diff(fun(x))  # производная
    trace = []
    trace.append(x_curr)
    x_prev = 0
    for i in range(num_iterations):
        x_new = x_curr - alpha * df_x.subs(x, x_curr) + betta*(x_curr - x_prev)
        trace.append(x_new)

        if abs(x_new - x_curr) < epsilon:
            return x_new, trace

        x_prev = x_curr
        x_curr = x_new

    return x_curr, trace
```
